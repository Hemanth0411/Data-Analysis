1) Instruction ( For Data Cleaning )
- Find all Null Values in the dataset. If there is any null value in any column, then fill it with the mean of that column.

```python
car.dtypes
```

```python
# prompt: delete multiple rows that have  null value in multiple coll

car = car.dropna(subset=['Make', 'Model', 'Type','Origin', 'DriveTrain','Invoice' ])
```

```python
car.shape
```

```python
car.isnull().sum()
```

```python
car['Cylinders'].fillna(car['Cylinders'].mean(), inplace = True)
```

```python
car.isnull().sum()
```
